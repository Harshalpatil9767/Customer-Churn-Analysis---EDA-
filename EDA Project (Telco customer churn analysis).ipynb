```python
# After analyzing the dataset we have found some blank spaces in "Total charges" coloumn

df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
def conv(value):
    if value == 1:
        return "Yes"
    else:
        return "No"
   
df['SeniorCitizen'] = df["SeniorCitizen"].apply(conv)
```

# for coloumn "Senior Citizen" converted 0 and 1 values of senior citizen to yes/no to make it easier to understand

```python
 # Dropping missing values 
df.dropna(inplace=True)

# drop customerID 
df.drop('customerID', axis=1, inplace=True)
```

```python
# applying loop check unique value without int or float in dataset

for col in df.columns:
    if df[col].dtype != 'int64' and df[col].dtype != 'float64':
        print(f'{col} : {df[col].unique()}')
```

```python
ax = sns.countplot(x = 'Churn', data = df)

ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Churn")
plt.show()
```

```python
plt.figure(figsize = (3,4))
gb = df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb['Churn'], labels = gb.index, autopct = "%1.2f%%")
plt.title("Percentage of Churned Customeres", fontsize = 10)
plt.show()
```

# from the given pie chart we can conclude that 26.54% of our customers have churned out.

# below are the reasons behind the churn

```python
plt.figure(figsize = (3,3))
sns.countplot(x = "gender", data = df, hue = "Churn")
plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x = "SeniorCitizen", data = df)
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Senior Citizen")
plt.show()
```

```python
total_counts = df.groupby('SeniorCitizen')['Churn'].value_counts(normalize=True).unstack() * 100

# Plot
fig, ax = plt.subplots(figsize=(4, 4))  # Adjust figsize for better visualization

# Plot the bars
total_counts.plot(kind='bar', stacked=True, ax=ax, color=['#1f77b4', '#ff7f0e'])  # Customize colors if desired

# Add percentage labels on the bars
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.text(x + width / 2, y + height / 2, f'{height:.1f}%', ha='center', va='center')

plt.title('Churn by Senior Citizen (Stacked Bar Chart)')
plt.xlabel('SeniorCitizen')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=0)
plt.legend(title='Churn', bbox_to_anchor = (0.9,0.9))  # Customize legend location

plt.show()
```

# we can see, comparatively a greater pecentage of people in senior citizen category have churned

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x = "Contract", data = df, hue = "Churn")
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Contract")
plt.show()
```

```python
# Key insights:

# Month-to-month contracts have the highest churn (orange bar) and also the highest customer count overall.

# One year and Two year contracts have significantly lower churn, suggesting that longer contracts help reduce churn.

# Most loyal customers (those who didn't churn) are found in longer-term contracts.

# This indicates that contract length is a strong factor influencing customer retention.
```

```python
columns = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 
           'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

# Number of columns for the subplot grid (you can change this)
n_cols = 3
n_rows = (len(columns) + n_cols - 1) // n_cols  # Calculate number of rows needed

# Create subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 4))  # Adjust figsize as needed

# Flatten the axes array for easy iteration (handles both 1D and 2D arrays)
axes = axes.flatten()

# Iterate over columns and plot count plots
for i, col in enumerate(columns):
    sns.countplot(x=col, data=df, ax=axes[i], hue = df["Churn"])
    axes[i].set_title(f'Count Plot of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')

# Remove empty subplots (if any)
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
#Service-wise Churn Analysis
#The chart presents multiple count plots showing how customer churn relates to different services.

#Key insights:

#PhoneService:
#Most customers have phone service. Churn is noticeably higher among them compared to those without.

#MultipleLines:
#Customers with multiple lines tend to churn more than those without. Those without phone service show the least churn.

#InternetService:
#Customers with Fiber optic internet have the highest churn rate, followed by DSL. Those with no internet service churn the least.

#OnlineSecurity, OnlineBackup, DeviceProtection, TechSupport:
#In all these categories, customers without the service are more likely to churn. This suggests that these value-added services help improve customer retention.

#StreamingTV & StreamingMovies:
#Churn is relatively even between customers who have these services and those who don’t. However, customers without internet service (and thus no streaming) show very low churn.

```


# Services that enhance customer experience or support (like security and backup services) appear to reduce churn. On the other hand, high-speed internet users, especially those on fiber optic, may be more likely to churn — potentially due to cost or service expectations.

```python
plt.figure(figsize = (6,4))
ax = sns.countplot(x = "PaymentMethod", data = df, hue = "Churn")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("Churned Customers by Payment Method")
plt.xticks(rotation = 45)
plt.show()
```

```python
# Churn Analysis by Payment Method
#The Above plot shows the distribution of churned and non-churned customers based on their payment methods.

# Key Insights:

# Electronic check users have the highest churn rate, with churn numbers almost equal to non-churn.

# Customers using automatic payment methods (bank transfer or credit card) show much lower churn rates.

# Mailed check users also churn less compared to electronic check users, but more than automatic methods.
```

# Automatic payment methods are associated with lower churn, suggesting that ease and automation in billing may contribute to customer retention.

```python
# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot for Monthly Charges
sns.kdeplot(data=df, x="MonthlyCharges", hue="Churn", fill=True, alpha=0.5, ax=axes[0])
axes[0].set_title('Density Plot of Monthly Charges by Churn Status')
axes[0].set_xlabel('Monthly Charges')
axes[0].set_ylabel('Density')

# Plot for Total Charges
sns.kdeplot(data=df, x="TotalCharges", hue="Churn", fill=True, alpha=0.5, ax=axes[1])
axes[1].set_title('Density Plot of Total Charges by Churn Status')
axes[1].set_xlabel('Total Charges')
axes[1].set_ylabel('Density')

plt.tight_layout()
plt.show()
```

```python
# High monthly charges correlate with churn, while low total charges (early tenure) indicate higher churn risk.

# Churners: high monthly, low total charges. Non-churners: spread across monthly, higher total charges.

# Suggestions:

# Early-tenure retention.
# Re-evaluate high-cost plans.
```
